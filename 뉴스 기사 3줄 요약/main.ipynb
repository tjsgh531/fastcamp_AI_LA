# base64 라이브러리 다루기

- 바이너리 파일(이미지, 영상)을 문자열 파일로 변환(ascii)

- 바이너리 파일을 6비트(2^6 = 64)로 나누어서 각 6비트의 수가 asciii문자열에 해당되는 문자로 변환

```python
import base64

string = 'Life is too short. We need Python !'
```

```python
# ascii 인코딩
bstring = string.encode('ascii')
print(bstring)
```

```python
# base64 인코딩
encoded = base64.b64encode(bstring)
print(encoded)
```

```python
# base64 디코딩
decode = base64.decodebytes(encoded)
print(decode) 
```

```python
# ascill 디코딩
decode_str = decode.decode('ascii')
print(decode_str)
```

### base64 사용해서 이미지 인코딩/디코딩

```python
import base64
path = "빵빵이.jpg"
```

```python
from PIL import Image
img = Image.open(path)
img
```

```python
# 바이너리 파일 읽기
with open(path, 'rb') as img:
    image = img.read()
```

```python
image
```

```python
import base64

# 이미지 파일을 base64로 인코딩
with open(path, 'rb') as img:
    data = img.read()
    encoded = base64.b64encode(data)
    print(encoded)
```

```python
# base64 디코딩
decoded = base64.decodebytes(encoded)
print(decoded)
```

```python
# 디코딩 데이터 이미지파일로 저장
file_path = "decoded.png"

with open(file_path, 'wb') as file:
    file.write(decoded)
```

# textwrap 라이브러리 사용하기

- 파이썬에서 문자열을 보기 좋은 형태로 정렬 또는 줄바꿈 하는데 사용할 수 있는 라이브러리

- 원하는 기이에 맞게 줄이기 : textwrap.shorten()

- 긴 문장 자르기 : textwrap.wrap()

- 긴 문장 줄바꿈 : textwrap.fill()


#### 데이터 가져오기 : 기사 데이터 크롤링

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

ChromeDriverManager().install()

driver = webdriver.Chrome()
```

```python
url = 'https://www.newscj.com/news/articleView.html?idxno=3140452'

driver.get(url)

title = driver.find_element(By.CLASS_NAME, 'heading').text
article = driver.find_element(By.XPATH, '//*[@id="article-view-content-div"]').text

data = title + ". " + article
data
```

#### textwrap.shorten()

너무 긴 문장을 중략해서 표현하고 싶을때

```python
import textwrap

textwrap.shorten(data, width=100, placeholder=" ...[중략]")
```

#### textwrap.wrap()

원하는 문자 수 만큼 나눠서 리스트 만들어 줌.

```python
wrapped_text = textwrap.wrap(data, width=40)
wrapped_text
```

```python
print('\n'.join(wrapped_text))
```

#### textwrap.fill()

textwrap.wrap() 과 '\n'.join() 을 합친 함수

두 기능을 동시에 구현

```python
fill_text = textwrap.fill(data, width=40)
print(fill_text)
```
